```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

url_train = "https://raw.githubusercontent.com/yogita5/test/main/classification_train%20(1).csv"
url_test = "https://raw.githubusercontent.com/yogita5/test/main/classification_test%20(1).csv"
df_train = pd.read_csv(url_train)
df_test = pd.read_csv(url_test)
data_train = df_train.values
data_test = df_test.values
```

```python
X_train = data_train[:,2:]
y_train = data_train[:,1]
X_test = data_test[:,1:]
```

```python
def distance(x1,x2):
  dist = np.sqrt(np.sum(np.square(x1-x2),axis = 1))
  return dist
```

```python
def KNN(X,X_train,y,k,i):
  neighbour_distance = []
  neighbour_distance = np.sqrt(np.sum(np.square(X-X_train),axis = 1))

  nearest_index = np.argpartition(neighbour_distance,k-1)
  y_nearest = y[nearest_index[:k]]
  
  y_pred = np.bincount(y_nearest).argmax()
 # print(f"Iteration: {i}     y-predicted: {y_pred}")
  return y_pred
```

```python
k = 4
y_pred = []
for i in range(X_test.shape[0]):
   y_pred.append(KNN(X_test[i],X_train,y_train,k,i))
print(y_pred)
```
